```python
import csv
import numpy as np
import matplotlib.pyplot as plt


def transformIntoFunction(filename):
    file = open(filename)
    csvreader = csv.reader(file,delimiter='\t')
    time = []
    position = []
    for row in csvreader:
        time.append(float(row[2].replace(',','.')))
        position.append(float(row[3].replace(',','.')))
        
    return time, position



def getArrays(filename):
    file = open(filename)
    csvreader = csv.reader(file,delimiter='\t')
    time = []
    position = []
    num = []
    for row in csvreader:
        time.append(float(row[0].replace(',','.')))
        position.append(float(row[1].replace(',','.')))
        num.append(float(row[2].replace(',','.')))
        
    return time, position, num



def trajectory(filename):
    file = open(filename)
    csvreader = csv.reader(file,delimiter='\t')
    time = []
    position = []
    for row in csvreader:
        time.append(float(row[0].replace(',','.')))
        position.append(float(row[1].replace(',','.')))
        
    return time, position





def getVelocityArray(filename):
    file = open(filename)
    csvreader = csv.reader(file,delimiter='\t')
    position = []
    for row in csvreader:

        position.append(float(row[2]))
        
    return position


def parseEscape(filename):
    file = open(filename)
    csvreader = csv.reader(file,delimiter='\t')
    time = []
    position = []
    for row in csvreader:
        time.append(float(row[0]))
        if((row[1]) == "true"): position.append("absorbed")
        elif((row[2]) == "true"): position.append("up")
        elif((row[3]) == "true"): position.append("down")
        elif((row[4]) == "true"): position.append("left")
        elif((row[5]) == "true"): position.append("right")
            
    return time, position

def getIndex(p,x):
    t = 0
    for xi in x:
        if p <= xi:
            return t
        t += 1
    return -1

def getItem(p,x):
    t = 0
    for xi in x:
        if p <= xi:
            return xi
        t += 1
    return -1

def getMean(x,y):
    toret = []
    for i in range(0,x):
        toappend = 0
        #n = size(y.values())
        for yi in y.keys():
            toappend += y[yi][i]/x
        
        toret.append(toappend)
        
    return toret


def getVariation(m):
    toret = []
    for mi in m:
        aux = np.sqrt((mi-m[0])**2)/m[0]
        toret.append(aux)
    
    return toret


def getScalarAvg(y):
    return getAvg(getVariation(y))

def getAvg(x):
    toret = 0
    for xi in x:
        toret += xi
    return toret/len(x)

def getStdDeviation(x):
    aux = 0
    xavg = getAvg(x)
    for xi in x:
        aux += (xi-xavg) * (xi-xavg)
    return np.sqrt(aux/len(x))
    

def countStrings(x,strg):
    toret = 0
    for xi in x:
        if(xi == strg): toret += 1
    return toret

def getArrayFromString(x,y,strg):
    toret = []
    for i in range(0,len(x)):
        if y[i] == strg:
            toret.append(x[i])
    return toret

@np.vectorize
def constant_function(x):
    return 0.75


```

```python
simid = [1,2,3,4,5] #5 simulaciones
percentages = [10,20,30,60,70,80]
rules = ["Rules3", "Rules2", "RulesOfLife"] #Las distintas reglas
dims = ["2D"]  
iterations = {}
max_distances = {}
cell_nums= {}

#Populo mis diccionarios con los arrays
for r in rules:
    iterations[r] = {}
    max_distances[r] = {}
    cell_nums[r] = {}
    for d in dims:
        iterations[r][d] = {}
        max_distances[r][d] = {}
        cell_nums[r][d] = {}
        
        for p in percentages:
            iterations[r][d][p] = {}
            max_distances[r][d][p] = {}
            cell_nums[r][d][p] = {}
            for i in simid:
                iterations[r][d][p][i], max_distances[r][d][p][i], cell_nums[r][d][p][i] = getArrays(((("../resources/"+d) + ((("/"+ r) + (("/" + str(i)) + ("/stats_" + d))) +"_")) +  str(p)) + ".txt")


```

```python
last_dists = {}
last_num_of_cells = {}
for r in rules:
    last_dists[r] = {}
    last_num_of_cells[r] = {}
    for d in dims:
        last_dists[r][d] = {}
        last_num_of_cells[r][d] = {}
        for p in percentages:
            last_dists[r][d][p] = []
            last_num_of_cells[r][d][p] = []
            for i in simid:
                last_dists[r][d][p].append(max_distances[r][d][p][i][-1])
                last_num_of_cells[r][d][p].append(cell_nums[r][d][p][i][-1])
                
dist_averages = {}
dist_stds = {}
cells_averages = {}
cells_stds = {}
for r in rules:
    cells_averages[r] = {}
    cells_stds[r] = {}
    dist_averages[r] = {}
    dist_stds[r] = {}
    for d in dims:
        cells_averages[r][d] = []
        cells_stds[r][d] = []
        dist_averages[r][d] = []
        dist_stds[r][d] = []
        for p in percentages:
            dist_averages[r][d].append(np.mean(last_dists[r][d][p]))
            dist_stds[r][d].append(np.std(last_dists[r][d][p]))
            cells_averages[r][d].append(np.mean(last_num_of_cells[r][d][p]))
            cells_stds[r][d].append(np.std(last_num_of_cells[r][d][p]))
            
            
            

    
```

```python


ax = plt.gca()




plt.xlabel("Porcentaje de celdas vivas iniciales",fontsize=12)
plt.ylabel("Radio final",fontsize=12)

ax.figure.set_size_inches(10,6)

plt.errorbar(percentages,dist_averages["Rules3"]["2D"],dist_stds["Rules3"]["2D"],fmt='.',capsize=2,capthick=1, ms = 12)
```
